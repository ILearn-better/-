```python
facePath=r'C:\Users\6\Anaconda3\Lib\site-packages\cv2\data\haarcascade_frontalface_alt2.xml'
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
__author__ = 'Seven'
import cv2 
def CatchPICFromVideo(catch_num, path_name):
    face_cascade = cv2.CascadeClassifier(facePath) 
    # haarcascade_frontalface_alt2
    camera = cv2.VideoCapture(0)#参数是0，表示打开笔记本的内置摄像头
    num = 0
    while True:
        ret, frame = camera.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.1, 5)
        for (x, y, w, h) in faces:
            img_name = f'{path_name}/{str(num)}.jpg'
            """
            path_name = '1'
            num = 2
            img_name = f'{path_name}/{str(num)}.jpg'
            img_name
            输出:'1/2.jpg'

            """
            # 以 f开头表示在字符串内支持大括号内的python 表达式
            image = frame[y:y + h, x:x + w]#提取方框内像素
            print(img_name)
            cv2.imwrite(img_name, image)#将image矩阵写入当前文件夹下的名为img_name图片文件
            num += 1#控制生成图片名字
            if num > catch_num:
                break
 
            # 画出矩形框圈出人脸
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
            # 显示捕捉了多少张人脸
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(frame, f'num:{str(num)}', (x + 30, y + 30), font, 1, (255, 0, 255), 4)
        if num > catch_num:
            break
        # 显示图像
        cv2.imshow('camera', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    camera.release()
    cv2.destroyAllWindows()
 
 
if __name__ == '__main__':
    CatchPICFromVideo(1000, './face_serve/1')
    CatchPICFromVideo(1000, './face_serve/0')    
```
